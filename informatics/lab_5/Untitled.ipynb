```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SPFB.RTS-12.18_180901_181231.csv')
```

```python
df['date'] = pd.to_datetime(df['<DATE>'])
```

```python
df_20 = df[(df['date'].dt.day == 20) & ((df['date'].dt.month == 9) | (df['date'].dt.month == 11) | (df['date'].dt.month == 12))]
df_22 = df[(df['date'].dt.day == 22) & (df['date'].dt.month == 10)]
```

```python
agg_df = pd.concat([df_20, df_22])
```

```python
agg_df['month_name'] = agg_df['date'].dt.month
```

```python
agg_df.groupby('month_name')[['<OPEN>', '<HIGH>', '<LOW>', '<CLOSE>']].boxplot(subplots=False, rot=15, fontsize=10, figsize=(14,10))
```
